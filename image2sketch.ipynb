## Procedure: 

> **Original Image → Gamma → Gray Scale → Invert → Blur → Invert → Pencil Sketch**



## Importing Libraries



```python
import numpy as np
from imutils.video import FPS
import imutils
import cv2
```

## Gamma Function

```python
def adjust_gamma(image, gamma=1.0):

   invGamma = 1.0 / gamma
   table = np.array([((i / 255.0) ** invGamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

   return cv2.LUT(image, table)
```

```python
#cap = cv2.VideoCapture("../image2sketch/Images/bugs1.mp4")
video = cv2.VideoCapture("../image2sketch/Images/bugs1.mp4")
frameCount = 0
frame_width = int(video.get(3))
frame_height = int(video.get(4))
size = (frame_width, frame_height)
result = cv2.VideoWriter('../image2sketch/Images/outputs/bugs1output.mp4', 
                         cv2.VideoWriter_fourcc(*'MP4V'),
                         10, size)
while True:

    ret, frame = video.read()
    frameCount += 1
    if frameCount % 3 == 0:
    
        frame = imutils.resize(frame, width = 450)
    

        # convert to sketch

        gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
        
        invert = cv2.bitwise_not(gray)

        blur = cv2.GaussianBlur(invert, (21, 21), 0, 0)
        
        invert_blur = cv2.bitwise_not(blur)

        div = cv2.divide(gray, invert_blur, scale=256)

        sketch = cv2.cvtColor(div,cv2.COLOR_GRAY2BGR)
    
        #result.write(sketch)


    # stack original and sketch frames

        display = np.vstack((sketch,frame))
    
    #write 
        result.write(display)

    # display the resulting frame

        cv2.imshow('pencil_sketch', display)
        
        key = cv2.waitKey(1)
        if key == ord('q'):
            break


# release capture

video.release()
result.release()
cv2.destroyAllWindows()



```

## Adjusting Gamma of Original Image

```python
##alpha = 1.5
##beta = 0
##contrast_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
##plt.axis('off')
##plt.imshow(contrast_img)
##plt.show()
gamma = 0.4
adj_img= adjust_gamma(img, gamma=gamma)
plt.axis('off')
plt.imshow(adj_img)
plt.show()
```

## Converting Original Image to Grey Scale

```python
gray_img = cv2.cvtColor(adj_img,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(gray_img)
plt.show()
```

## Inverting the Gray Image

```python
#inv_gray_img = 255 - gray_img
inv_gray_img = cv2.bitwise_not(gray_img)
plt.axis('off')
plt.imshow(inv_gray_img)
plt.show()
```

## Blurring the inverted gray image by Gausian Function

```python
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
plt.axis('off')
plt.imshow(blur_img)
plt.show()
```

```python
#inv_blur_img = 255 - blur_img
inv_blur_img = cv2.bitwise_not(blur_img)
plt.axis('off')
plt.imshow(inv_blur_img)
plt.show()
```

## Creating Pencil Sketch image

```python
pencil_img = cv2.divide(gray_img,inv_blur_img,scale = 256.0)
plt.axis('off')
plt.imshow(pencil_img)
plt.show()
```
