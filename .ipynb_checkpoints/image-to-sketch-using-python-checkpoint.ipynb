## Procedure: 

> **Original Image → Gamma → Gray Scale → Invert → Blur → Invert → Pencil Sketch**



## Importing Libraries



```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def adjust_gamma(image, gamma=1.0):

   invGamma = 1.0 / gamma
   table = np.array([((i / 255.0) ** invGamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

   return cv2.LUT(image, table)
```

## Reading The Input Image 

```python
img = cv2.imread("../Downloads/image2sketch/Images/input_image.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(img)
plt.show()
```

## Increasing Contrast of Original Image

```python
##alpha = 1.5
##beta = 0
##contrast_img = cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
##plt.axis('off')
##plt.imshow(contrast_img)
##plt.show()
gamma = 0.4
adj_img= adjust_gamma(img, gamma=gamma)
plt.axis('off')
plt.imshow(adj_img)
plt.show()
```

## Converting Original Image to Grey Scale

```python
gray_img = cv2.cvtColor(adj_img,cv2.COLOR_BGR2GRAY)
plt.axis('off')
plt.imshow(gray_img)
plt.show()
```

## Inverting the Gray Image

```python
#inv_gray_img = 255 - gray_img
inv_gray_img = cv2.bitwise_not(gray_img)
plt.axis('off')
plt.imshow(inv_gray_img)
plt.show()
```

## Blurring the inverted gray image by Gausian Function

```python
blur_img = cv2.GaussianBlur(inv_gray_img,(21,21),0)
plt.axis('off')
plt.imshow(blur_img)
plt.show()
```

```python
#inv_blur_img = 255 - blur_img
inv_blur_img = cv2.bitwise_not(blur_img)
plt.axis('off')
plt.imshow(inv_blur_img)
plt.show()
```

## Creating Pencil Sketch image

```python
pencil_img = cv2.divide(gray_img,inv_blur_img,scale = 256.0)
plt.axis('off')
plt.imshow(pencil_img)
plt.show()
```

```python
##multi_img = cv2.multiply(pencil_img, gray_img)
##plt.axis('off')
##plt.imshow(multi_img)
##plt.show()
```

```python
cv2.imwrite("../Downloads/image2sketch/Images/output_img.png",pencil_img)
```

```python
plt.figure(figsize=(14,8))

plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
##multi_img =cv2.multiply(pencil_img, img)
img_sketch=cv2.cvtColor(pencil_img, cv2.COLOR_BGR2RGB)
plt.imshow(img_sketch)
plt.axis('off')
plt.show()
```
